```python
def fentropy2(col_data, col_name):
    r = pd.DataFrame(columns = ['value', 'entropy', 'weight'])
    len_tot = len(col_data)
    gain = 0
    for i in col_data[col_name].unique():
        x = col_data[col_data[col_name] == i]
        col_weight = len(x)
        e1 = len(x[x['survived'] == 1])
        e2 = len(x[x['survived'] == 0])
        entr = entropy(e1, e2)
        idx = len(r) + 1
        r.loc[idx] = [i,entr,col_weight]
        gain = gain + (entr*col_weight/len_tot)
    
    return(r,gain)
```

```python
def split(data, cols):
    final_entropy  = pd.DataFrame(columns = ['split_value', 'entropy', 'column'])
    for col in cols:
        for i in data[col].unique():
            data['class'] = np.where(data[col] > i, 1,0)
            
            df, ent = fentropy2(data, 'class')
            idx = len(final_entropy) + 1
            final_entropy.loc[idx] = [i, ent, col]
    return(final_entropy[final_entropy['entropy'] == final_entropy['entropy'].min()])
```

```python
def pfind(data):
    if(len(data['survived'].unique()) > 0):
        p1 = len(data[data['survived'] == 1])/len(data)*100
        p0 = len(data[data['survived'] == 0])/len(data)*100
    else:   
        if(data['survived'].unique() == 0):
            p0 = 100
            p1 = 0
        else:
            p1 = 100
            p0 = 0

    return(p1, p0)
```

```python
tree = pd.DataFrame(columns = ['path', 'data_length', 'sur', 'not_sur', 'columnOfSplit', 'thresholdOfSplit'])
path = 'root'
def tree_maker(data, path):
    cols = data.columns.drop(['survived'])
    re = split(data, cols)
    if(np.shape(re)[0]>1):
        return(1)
    per_sur, per_nsur = pfind(data)
    tc = re['column'].item()
    sp = re['split_value'].item()
    data_left = data[ data[tc] > sp]
    data_right = data.drop(index = data_left.index)
    idx = len(tree) + 1
    tree.loc[idx] = [path, len(data),per_sur, per_nsur, tc, sp]
    if(per_sur == 0 or per_nsur == 0):
        idx = len(tree) + 1
        pathleaf = path + 'leaf'
        tree.loc[idx] = [pathleaf, len(data),per_sur, per_nsur, tc, sp]
        return(1)
    else:
        pathl = path + 'Left'
        pathr = path + 'Right'
        tree_maker(data_left, pathl)
        tree_maker(data_right, pathr)  

x = tree_maker(d2f, path)
```

```python
def accuracy_finder(data):
    tp = 0
    fp = 0
    tn = 0
    fn = 0
    tp = np.where(data['survived'] == data['predictions'], np.where(data['predictions']==1,1,0),0)
    fp = np.where(data['survived'] != data['predictions'], np.where(data['predictions']==1,1,0),0)
    tn = np.where(data['survived'] == data['predictions'], np.where(data['predictions']==0,1,0),0)
    fn = np.where(data['survived'] != data['predictions'], np.where(data['predictions']==0,1,0),0)
    precision = tp.sum()/(tp.sum()+fp.sum())
    recall = tp.sum()/(tp.sum()+fn.sum())
    fsc = 2/((1/precision) + (1/recall))
    accuracy = (tp.sum()+tn.sum())/(tp.sum()+tn.sum()+fp.sum()+fn.sum())
    return(accuracy,fsc,recall,precision)
```

```python
ntree = pd.DataFrame(columns = ['path', 'data_length', 'sur', 'not_sur', 'columnOfSplit', 'thresholdOfSplit'])
path = 'root'
def ntree_maker(data, path):
    cols = data.columns.drop(['survived'])
    re = split(data, cols)
    if(np.shape(re)[0]>1):
        return(1)
    per_sur, per_nsur = pfind(data)
    tc = re['column'].item()
    sp = re['split_value'].item()
    data_left = data[ data[tc] > sp]
    data_right = data.drop(index = data_left.index)
    idx = len(ntree) + 1
    ntree.loc[idx] = [path, len(data),per_sur, per_nsur, tc, sp]
    if(per_sur == 0 or per_nsur == 0):
        idx = len(ntree) + 1
        pathleaf = path + 'leaf'
        ntree.loc[idx] = [pathleaf, len(data),per_sur, per_nsur, tc, sp]
        return(1)
    else:
        pathl = path + 'Left'
        pathr = path + 'Right'
        ntree_maker(data_left, pathl)
        ntree_maker(data_right, pathr)  
```
